## If, elif, else statements example: Roller Coaster Height Checker

Before we start the exercises, let's look at an example program that determines whether someone is tall enough to ride a roller coaster.

**Requirements:**
- Get the user's height
- If height is 0 or less, print an error message
- If height is below minimum (100cm), tell them they cannot ride
- If height is between 100cm and 160cm, tell them they can ride with an adult
- If height is 160cm or above, tell them they can ride alone

**Example solution:**

````python
#Roller Coaster Height Checker
MIN_HEIGHT = 100
RIDE_ALONE_WHEN_HEIGHT_ABOVE = 160

height = int(input("Enter_in_your_height_in_cm:"))
            
if height <= 0:
    print("ERROR MESSAGE")

elif height<MIN_HEIGHT:
    print("You cannot ride")

elif height<RIDE_ALONE_WHEN_HEIGHT_ABOVE:
    print("You can ride with an adult")

elif height>RIDE_ALONE_WHEN_HEIGHT_ABOVE:
    print("You can ride alone")
````

## Exercise 1: Addition

Create a program that adds together the number of students and tutors in a class.

**Requirements:**
- Ask for two integers with the message: `"Please enter the number of students and tutors: "`
- Display the sum in the format: `students + tutors = sum`

**Examples:**
```
Please enter the number of students and tutors: 2 5
2 + 5 = 7
```

```
Please enter the number of students and tutors: -1 5
-1 + 5 = 4
```

**Hint:** Use `input().split()` to read two numbers on one line, then convert them to integers.

````python
students_and_tutors = input("Please enter the number of students and number of tutors")
print(students_and_tutors)
students, tutors = students_and_tutors.split()
print(students)
print(tutors)

total_people = int(students)+int(tutors)
print(total_people)
````

## Exercise 4: Leap Year

**This is a pair programming exercise!**

Write a program that determines whether a given year is a leap year.

**Leap Year Rules:**
- A year is a leap year if it is divisible by 4
- EXCEPT years divisible by 100 are NOT leap years
- UNLESS they are also divisible by 400 (then they ARE leap years)

**Examples:**
- 2016 is a leap year (divisible by 4)
- 2000 is a leap year (divisible by 400)
- 1900 is NOT a leap year (divisible by 100 but not 400)
- 2017 is NOT a leap year (not divisible by 4)

**Sample runs:**
```
Enter year: 2016
2016 is a leap year.
```

```
Enter year: 2017
2017 is not a leap year.
```

**Hint:** Use the modulo operator `%` to check divisibility

````python
# Exercise 4: Leap Year
# Write your code here
year_variable = int(input("Enter the year"))

if year_variable % 400 ==0:
    print(f"{year_variable} is a leap year")

elif year_variable% 100 ==0:
    print(f"{year_variable} is not a leap year")

elif year_variable %4 == 0: 
    print(f"{year_variable} is a leap year")









````
